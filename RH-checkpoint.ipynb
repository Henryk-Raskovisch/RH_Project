```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Salary_Data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar = False);
```

```python
df.describe().round(2)
```

```python
plt.figure(figsize = (10, 6))
plt.title("salary Analysis")
sns.kdeplot(df["Salary"], color = "green");
```

```python
plt.figure(figsize = (10, 6))
plt.title("Experience Analysis")
sns.kdeplot(df["YearsExperience"], color = "blue");
```

```python
plt.figure(figsize = (10, 5))
sns.boxplot(x = df["Salary"]);
```

```python
plt.figure(figsize = (10, 5 ))
sns.boxplot(x = df["YearsExperience"]);
```

```python
#distribution
plt.figure(figsize = (10, 5))
sns.scatterplot(data = df, x = "Salary", y = "YearsExperience");
```

```python
plt.figure(figsize = (10, 5))
sns.regplot(data = df, x = "Salary", y = "YearsExperience");
```

```python
#correlation
correlation = df.corr("pearson")
plt.figure(figsize = (10, 5))
sns.heatmap(correlation, annot = True, cbar = True);
```
